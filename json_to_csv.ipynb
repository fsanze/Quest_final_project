```python
import pandas as pd
import json

pd.set_option('display.max_colwidth', None)
```

```python
#cargamos el JSON
with open("data3.json", "r") as file:
    data = json.load(file)
```

```python
ticker_list = []
for i in range(len(data)):
    ticker_list.append(data[i]["Meta Data"]["2. Symbol"])
ticker_list
```

```python
#Creamos una lista vacia para ir guardando la información ya con la estructura que queremos
all_data = []

#Bucle para ir pasando por los datos de cada valor
for element in data:
    ticker = element["Meta Data"]["2. Symbol"]
    
    for key, value in element["Time Series (Daily)"].items():
        #creamos un diccionario con las columnas que vamos a querer en el df y se lo añadimos a la lista 
        all_data.append({
            "Date": key,
            "Symbol": ticker,
            "Open": round(float(element["Time Series (Daily)"][key]["1. open"]),2),
            "High": round(float(element["Time Series (Daily)"][key]["2. high"]),2),
            "Low": round(float(element["Time Series (Daily)"][key]["3. low"]),2),
            "Close": round(float(element["Time Series (Daily)"][key]["4. close"]),2),
            "Volume": int(element["Time Series (Daily)"][key]["5. volume"]),
            })

#Formamos el df y lo omprobamos
df = pd.DataFrame(all_data)
df
```

```python
#Guardamos los datos en CSV
df.to_csv("data3.csv", index=False)
```

```python
#Cargamos los CSV, hacemos el merge por el ID y comprobamos
df1 = pd.read_csv("data1.csv")
df2 = pd.read_csv("data2.csv")
df3 = pd.read_csv("data3.csv")
df = pd.concat([df1, df2, df3], ignore_index=True)
df
```
